```python
def transcripts_done(transcript_folder='news_transcripts'):
    # 1. Setup
    folder_path = os.path.join(BASE_PATH, transcript_folder) # Change this to your folder
    video_ids = []

    # 2. Iterate through files
    for filename in os.listdir(folder_path):
        if filename.endswith('.json'):
            file_path = os.path.join(folder_path, filename)
            
            # 3. Open and parse
            with open(file_path, 'r', encoding='utf-8') as f:
                try:
                    data = json.load(f)
                    
                    # 4. Extract field (using .get() to avoid crashing if field is missing)
                    v_id = data.get('video_id')
                    if v_id is not None:
                        video_ids = video_ids + [v_id]
                except json.JSONDecodeError:
                    print(f"Error: Could not parse {filename}")
    return video_ids
```

```python
def get_yt_id(video_url):
    pattern = r"(?:v=|\/|be\/)([0-9A-Za-z_-]{11})"
    match = re.search(pattern, video_url)
    if match:
        return match.group(1)
    return None
```

```python
transcripts = transcripts_done()
```

```python
def get_json(video_id, transcripts_list = transcripts, result_folder = 'news_transcripts/'):
    if video_id in transcripts_list and video_id != '249z9TVsvho':
        return

    video_url = f"https://www.youtube.com/watch?v={video_id}"

    # ========
    # GET INFO OF THE VIDEO 
    # ========
    ydl_opts = {
        "quiet": True,
        "skip_download": True,
    }

    with YoutubeDL(ydl_opts) as ydl:
        info = ydl.extract_info(video_url, download=False)
    
    title = info.get("title")
    channel = info.get("channel")
    uploader = info.get("uploader")
    upload_date = info.get("upload_date")
    duration = info.get("duration")

    print("Video's title:", title)
    print("Channel:", channel)
    print("Uploader:", uploader)
    print("Upload date:", upload_date)
    print("Duration:", duration, "s")

    # ========
    # CALL API
    # ========

    api = YouTubeTranscriptApi()
    transcript_list = api.list(video_id)
    transcript = transcript_list.find_generated_transcript(['hu'])
    data = transcript.fetch()
    full_text = " ".join(entry.text for entry in data)
    data = {
        'title':title,
        'channel':channel,
        'upload_date':upload_date,
        'duration':duration,
        'raw_text':full_text
    }

    return data

```

```python
video_ids_list = open(IDS_FILE, encoding='utf-8').read().splitlines()
```

```python
transcripts = transcripts_done()

for video_id in tqdm(video_ids_list):
    if video_id in transcripts:
        print('The current video id is already transcripted, move to the next one.')
    else:
        # call the function
        try:
            json_data = get_json(video_id)

            # extract information
            channel = json_data.get('channel', 'UnknownChannel')
            upload_date = json_data.get('upload_date', '00000000')
            clean_channel = "".join(c for c in channel if c.isalnum() or c in (' ', '_')).strip()
            file_name = f"{clean_channel}_{upload_date}.json"

            # save data
            output_path = os.path.join(BASE_PATH, 'news_transcripts', file_name)
            with open(output_path, 'w', encoding='utf-8') as f:
                json.dump(json_data, f, indent=4, ensure_ascii=False)
            print(f"Successfully saved: {file_name}. Now the code is waiting to the next one.")
            time.sleep(random.uniform(30, 60))
            print('The waiting is over, moving to the next video.')
        except Exception as e:
            print(f'There was an error at {video_id}.')
            print(f'The error is the following: {e}')
```
